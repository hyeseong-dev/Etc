```python
import random
```

```python
def lotto_random():
    # num_list에 1부터 45까지 값을 넣는 방법 2가지            
    # 1. 컴프리헨션           
    # 2. for문과 append를 이용한 삽입방법         


    # 컴프리헨션으로 삽입하기 
    # num_list = [i for i in range(1,46)]


    # for문과 append를 이용함
    num_list=[]
    for i in range(1,46):
        num_list.append(i)
    # print(num_list)

    # 리스트에서 무작위로 값을 뽑는 방법은:   
    # 1. random모듈의 randrange()를 이용 할 수 있음.  


    win_num_list = []   # 0~45까지의 기존 num_list에서 끄집어낸걸 저장하는 리스트
    for i in range(0,6):# 로또번호 6번 뽑아야지?!
        r_seed = random.randrange(0,len(num_list)) # num_list의 index범위가 0~44까지와 동일함
        win_num = num_list.pop(r_seed) # 수를 끄집어내버림.len역시 줄어듬.
        win_num_list.append(win_num) 
    win_num_list.sort() 
    print(win_num_list)

    lotto = [3,4,16,27,38,40]
    lotto_bonus = 20

    point = 0
    for i in range(0,len(win_num_list)):
        for j in range(0,len(lotto)):
            if win_num_list[i] == lotto[i]:
                point +=1

    point2 = 0
    for i in range(0,len(win_num_list)):
        if win_num_list[i] == lotto_bonus:
            point2 = 1
    # print(point, point2)

    # 1등 : 6개   
    # 2등 : 5개 + 보너스 : 1개    
    # 3등 : 5개 + 보너스 : 0개    
    # 4등 : 4개   
    # 5등 : 3개   

    if point == 3:
        print('5등 ')
    elif point == 4:
        print('4등')
    elif (point == 5) and (point2 == 0):
        print('3등')
    elif (point == 5) and (point2 == 1):
        print('2등')
    elif (point == 5) and (point2 == 1):
        print('1등')
    else: print('꽝')
```

```python
for i in range(0,300):
    lotto_random()
```
